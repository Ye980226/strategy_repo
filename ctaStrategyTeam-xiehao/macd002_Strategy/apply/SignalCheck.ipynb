# 实盘信号检测

```python
with open("vt_20181110.log","r",encoding="utf-8") as f:
    content = f.readlines()
data = []
for logRow in content:
    if logRow.find("CTA_STRATEGY	macd002_Strategy:eos_this_week:OKEX_WUon_1min_bar")>=0:
        result = logRow.split("CTA_STRATEGY	macd002_Strategy:eos_this_week:OKEX_WUon_1min_bar")[-1]
        data.append(result)
```

```python
import pandas as pd
from datetime import datetime
```

```python
checkDict = {'datetime':[], 'trend':[], 'cross':[], 'mean':[], 'longStop':[], 'longexit':[], 'shortStop':[], 'shortexit':[]}
for row in data:
    dateTime = row[row.find('time')+5:row.find('date')+26]
    rigthTime = datetime.strptime(dateTime, "%Y-%m-%d %H:%M:%S")
    checkDict['datetime'].append(rigthTime)
    checkDict['trend'].append((row[row.find('trend')+6:].split(',')[0]))
    checkDict['cross'].append((row[row.find('cross')+6:].split(',')[0]))
    checkDict['mean'].append((row[row.find('mean')+5:].split(',')[0]))
    checkDict['longStop'].append((row[row.find('mean')+9:].split(',')[0]))
    checkDict['longexit'].append((row[row.find('mean')+9:].split(',')[0]))
    checkDict['shortStop'].append((row[row.find('mean')+9:].split(',')[0]))
    checkDict['shortexit'].append((row[row.find('mean')+9:].split('/n')[0]))
```

```python
realCheck = pd.DataFrame(checkDict)
```

```python
realCheckDf = realCheck.set_index('datetime')
```

# 回测信号检查

```python
checkBtDict = {'datetime':[], 'trend':[], 'cross':[], 'mean':[], 'longStop':[], 'longexit':[], 'shortStop':[], 'shortexit':[]}
```

```python
import pandas as pd

btLog = pd.read_csv('BacktestLog.csv')
btData = []

for i, value in btLog.iterrows():
    if value.values[0].find("macd002_Strategy:eos_this_week:OKEX_WUon_1min_bar")>=0:
        result = value.values[0].split("macd002_Strategy:eos_this_week:OKEX_WUon_1min_bar")[-1]
        btData.append(result)
```

```python
from datetime import datetime

for row in btData:
    dateTime = row[row.find('time')+5:25]
    rigthTime = datetime.strptime(dateTime, "%Y-%m-%d %H:%M:%S")
    checkBtDict['datetime'].append(rigthTime)
    checkBtDict['trend'].append((row[row.find('trend')+6:].split(',')[0]))
    checkBtDict['cross'].append((row[row.find('cross')+6:].split(',')[0]))
    checkBtDict['mean'].append((row[row.find('mean')+5:].split(',')[0]))
    checkBtDict['longStop'].append((row[row.find('mean')+9:].split(',')[0]))
    checkBtDict['longexit'].append((row[row.find('mean')+9:].split(',')[0]))
    checkBtDict['shortStop'].append((row[row.find('mean')+9:].split(',')[0]))
    checkBtDict['shortexit'].append((row[row.find('mean')+9:].split('/n')[0]))
```

```python
checkBtData = pd.DataFrame(checkBtDict)
```

```python
checkBtDd = checkBtData.set_index('datetime')
```

```python
FinalDf = pd.concat([checkBtDd, realCheckDf], axis=1).dropna()
```
