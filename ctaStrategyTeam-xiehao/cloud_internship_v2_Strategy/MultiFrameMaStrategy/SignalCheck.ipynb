# 实盘信号检测

```python
with open("vt_20181023.log","r",encoding="utf-8") as f:
    content = f.readlines()
data = []
for logRow in content:
    if logRow.find("CTA_STRATEGY	ARPSYMaStrategy:on60MinBar")>=0:
        result = logRow.split("CTA_STRATEGY	ARPSYMaStrategy:on60MinBar")[-1]
        data.append(result)
```

```python
import pandas as pd
from datetime import datetime
```

```python
checkDict = {'datetime':[],'score-1':[],'score-2':[],'crossOver':[],'crossBelow':[],'maTrend':[]}
for row in data:
    dateTime = row[row.find('time')+5:29]
    rigthTime = datetime.strptime(dateTime, "%Y-%m-%d %H:%M:%S")
    checkDict['datetime'].append(rigthTime)
    checkDict['score-1'].append(row[row.find('score-1')+8:].split(',')[0])
    checkDict['score-2'].append(row[row.find('score-2')+8:].split(',')[0])
    checkDict['crossOver'].append(row[row.find('crossOver')+10])
    checkDict['crossBelow'].append(row[row.find("crossBelow")+11])
    checkDict['maTrend'].append(row[row.find("maTrend")+8])
```

```python
realCheck = pd.DataFrame(checkDict)
```

```python
realCheckDf = realCheck.set_index('datetime')
```

```python
realCheckDf
```

# 回测信号检查

```python
checkBtDict = {'datetime':[],'CurSignal':[],'maTrend':[],'CurvolSignal':[],'Stoptracking':[],'maStop':[],'longStop':[],'longProfit':[],'shortStop':[],'shortProfit':[]}
```

```python
import pandas as pd

btLog = pd.read_csv('BacktestLog.csv')
btData = []

for i, value in btLog.iterrows():
    if value.values[0].find("Overshoot:on1MinBar")>=0:
        result = value.values[0].split("Overshoot:on1MinBar")[-1]
        btData.append(result)
```

```python
result
```

```python
result[result.find('maStop')+7:]
```

```python
from datetime import datetime

for row in btData:
    dateTime = row[row.find('time')+5:29]
    rigthTime = datetime.strptime(dateTime, "%Y-%m-%d %H:%M:%S")
    checkBtDict['datetime'].append(rigthTime)
    checkBtDict['CurSignal'].append(row[row.find('CurSignal')+10:].split(',')[0])
    checkBtDict['maTrend'].append(row[row.find('maTrend')+8:].split(',')[0])
    checkBtDict['CurvolSignal'].append(row[row.find('CurvolSignal')+13:].split(',')[0])
    checkBtDict['Stoptracking'].append(row[row.find("Stoptracking")+13:].split(',')[0])
    checkBtDict['maStop'].append(row[row.find("maStop")+7:].split(',')[0])
    checkBtDict['longStop'].append(row[row.find("longStop")+9:].split(',')[0])
    checkBtDict['shortStop'].append(row[row.find("shortStop")+10:].split(',')[0])
    checkBtDict['shortProfit'].append(row[row.find("shortProfit")+12:].split('\'')[0])
```

```python
checkBtData = pd.DataFrame(checkBtDict)
```

```python
checkBtDd = checkBtData.set_index('datetime')
```

```python
checkBtDd
```
